# geom_point()

```r
donnee=iris
head(iris)
```

```r
ggplot(data=donnee,aes(x=Sepal.Length,y=Sepal.Width,size=Petal.Width))+geom_point(color="blue")+theme_classic()
```

![](https://www.datanovia.com/en/wp-content/uploads/dn-tutorials/ggplot2/figures/030-ggplot-point-shapes-r-pch-list-showing-different-point-shapes-in-rstudio-1.png)

```r
ggplot(data=donnee,aes(x=Sepal.Length,y=Sepal.Width))+geom_point(size=5,shape=1,fill="orange")+theme_classic()
```

# geom_histogram()

```r
ggplot(data=donnee,aes(x=Sepal.Width))+geom_histogram(color="blue",fill="orange")+theme_bw()
```

# geom_col()

```r
library(dplyr)
df=donnee %>% group_by(Species) %>% summarise_all(mean) 

g=ggplot(df,aes(x=Species,y=Sepal.Width,fill=Species))+geom_col()+theme_classic()
g
```

# geom_density()

```r
ggplot(data=donnee,aes(x=Sepal.Width))+geom_density()
```

```r
t=ggplot(data=donnee,aes(y=Sepal.Width,x=Species,fill=Species))+geom_boxplot()+theme_classic()
```

# Personaliser le graphique avec theme(),labs()

```r
g
```

```r
g=g+labs(x="Voici",y="",title="C'est mon titre",subtitle="Sous titre",caption="Source")
g
```

```r
g+theme(

plot.title=element_text(color="#E67E22"),

)
```

```r
g
```

```r
?element_line
```

```r
g+theme(axis.line =element_blank())
```

```r
g+theme(axis.line.x = element_blank())
```

```r
g+theme(axis.line.y = element_line(color = "blue"),
       axis.line.x=element_line(color="orange"))
```

```r
g+theme(axis.ticks.x = element_blank())
```

```r
g+theme(axis.ticks.x =element_line(color = "blue",size=5),
       axis.ticks.y =element_line(color = "orange",size=5)
       )
```

```r
t+theme(axis.text.y = element_text(angle=90,
                                  color="blue",
                                  size=20))
```

```r
g+theme(legend.position = "none")
```
